```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
```

```python
t = np.arange(-5.0, 5.0, 0.1)
z = np.zeros(t.size)

plt.figure()
plt.subplot(2,1,1)
plt.plot(t, np.tanh(t))
plt.title('Hyperbolic Tangent (tanh)')
plt.axis('off')

plt.subplot(2,1,2)
plt.plot(t, np.amax([t,z], axis=0))
plt.title('Rectified Linear Unit (relu)')
plt.axis('off')
plt.savefig('activation_funs.eps')
```
